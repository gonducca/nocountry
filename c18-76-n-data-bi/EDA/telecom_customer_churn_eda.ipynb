```python
import pandas as pd
from ydata_profiling import ProfileReport

df = pd.read_csv("../Data/telecom_customer_churn.csv")
df.head()
ProfileReport(df)
```

```python
from sklearn.preprocessing import OneHotEncoder


# Crear una columna combinada de Zip Code, Latitude, y Longitude
df['Geolocation'] = df.apply(lambda row: f"{row['Zip Code']}_{row['Latitude']}_{row['Longitude']}", axis=1)

# Codificar la columna Geolocation
geo_encoder = OneHotEncoder()
geo_encoded = geo_encoder.fit_transform(df[['Geolocation']]).toarray()
geo_encoded_df = pd.DataFrame(geo_encoded, columns=geo_encoder.get_feature_names_out(['Geolocation']))

#Eliminar columnas originales


df.drop(columns=['Zip Code', 'Latitude', 'Longitude'], inplace=True)
df.head()
```
